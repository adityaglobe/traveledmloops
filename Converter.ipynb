```python
from nbconvert import PythonExporter
import nbformat
```

```python
with open("Travel EDA.ipynb","r",encoding='utf8') as file:
    NC=nbformat.read(file,as_version=4)
```

```python
PE=PythonExporter()
```

```python
(script,_)=PE.from_notebook_node(NC)
with open("modelfile.py",'w') as sf:
    sf.write(script)
```
